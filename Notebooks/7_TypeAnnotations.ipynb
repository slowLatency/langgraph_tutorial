## **TYPE ANNOTATIONS**

#### **TypeDict**

```python
from typing import TypedDict

class Movie(TypedDict):
    name: str
    release_date: str

movie = Movie(name="Avengers", release_date="17th April 1999")

movie
```

##### **Union**

```python
from typing import Union

def print_item(x: Union[int,float]):
    print(x)

print_item(5)
```

#### **Optional**

```python
from typing import Optional

def print_item(x: Optional[str]):
    if x is None:
        print("None")
    else:
        print(x)
    
(print_item(5))

```

#### **Any**

```python
from typing import Any

def print_item(x: Any):
    print(x)

print_item("Hello")
```

#### **LambdaFunction**

```python
nums = [1,2,3,4]

square = list(map(lambda x: x*x, nums))
print(square)
```
