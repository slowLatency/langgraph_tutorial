```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
import random

class GraphSchema(TypedDict):
    numbers: list
    counter: int


def random_node(state):
    state['numbers'].append(random.randint(0,10))
    state['counter'] += 1

    return state

def loop(state):
    if state['counter'] < 5:
        return "loop"
    else:
        return "end"

loop_graph = StateGraph(GraphSchema)
loop_graph.add_node("random", random_node)
loop_graph.add_edge(START, "random")
loop_graph.add_conditional_edges("random", loop, {"loop":"random", "end":END})

loop_graph_compiled = loop_graph.compile()

loop_graph_compiled.invoke(
    {
        'numbers': [],
        'counter':0
    }
)
```
