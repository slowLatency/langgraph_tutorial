```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

class GraphSchema(TypedDict):
    messages: str

def hello_world(state: GraphSchema)->GraphSchema:
    state['messages'] = "Hey " + state['messages'] + ". How are you doing?"
    return state


hello_world_graph = StateGraph(GraphSchema)
hello_world_graph.add_node('HelloWorld', hello_world)
hello_world_graph.add_edge(START, 'HelloWorld')
hello_world_graph.add_edge('HelloWorld', END)

hello_world_graph_compiled = hello_world_graph.compile()
hello_world_graph_compiled.invoke({"messages":"Adib"})
```
