```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

class GraphSchema(TypedDict):
    number1: int
    number2: int
    number3: int
    number4: int
    operation1: str
    operation2: str
    result1: int
    result2: int

def adder1(state):
    state['result1'] = state['number1'] + state['number2']
    return state

def subtractor1(state):
    state['result1'] = state['number1'] - state['number2']
    return state

def adder2(state):
    state['result2'] = state['number3'] + state['number4']
    return state

def subtractor2(state):
    state['result2'] = state['number3'] - state['number4']
    return state

def router1(state):
    if state['operation1'] == '+':
        return "add"
    else:
        return "subtract"

def router2(state):
    if state['operation2'] == '+':
        return "add"
    else:
        return "subtract"

conditional_graph = StateGraph(GraphSchema)
conditional_graph.add_node("router1", lambda state: state)
conditional_graph.add_node("adder1", adder1)
conditional_graph.add_node("subtractor1", subtractor1)
conditional_graph.add_node("router2", lambda state: state)
conditional_graph.add_node("adder2", adder2)
conditional_graph.add_node("subtractor2", subtractor2)
conditional_graph.add_edge(START, "router1")
conditional_graph.add_conditional_edges("router1", router1, {"add":"adder1", "subtract":"subtractor1"})
conditional_graph.add_edge("adder1", "router2")
conditional_graph.add_edge("subtractor1", "router2")
conditional_graph.add_conditional_edges("router2", router2, {"add":"adder2", "subtract":"subtractor2"})
conditional_graph.add_edge("adder2", END)
conditional_graph.add_edge("subtractor2", END)

conditional_graph_compiled = conditional_graph.compile()
conditional_graph_compiled
conditional_graph_compiled.invoke(
    {
        "number1":5,
        "number2": 10,
        "operation1":"+",
        "number3":10,
        "number4":6,
        "operation2":"-"
    }
)
```
