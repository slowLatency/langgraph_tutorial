### **Messages**

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

class GraphSchema(TypedDict):
    messages: list

llm = ChatOpenAI(model='gpt-5-mini', temperature=0)

def greet(state: GraphSchema)->GraphSchema:
    messages = state['messages']
    state['messages'].append(AIMessage(content=llm.invoke(messages).content))

    return state

my_message_graph = StateGraph(GraphSchema)
my_message_graph.add_node('greet', greet)
my_message_graph.add_edge(START, 'greet')
my_message_graph.add_edge('greet', END)

my_message_graph_compiled = my_message_graph.compile()
my_message_graph_compiled.invoke(
    {
        'messages':[
            SystemMessage(content = 'You are a greeting assistant.'),
            HumanMessage(content = 'Hi I am Adib, How are you?')
        ]
    }
)
```
