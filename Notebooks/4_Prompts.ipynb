### **Prompts**

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI(model='gpt-5', temperature=0)

prompt_tempelate = ChatPromptTemplate.from_messages(
    [
        ('system', 'You are a cool assistant, whose age is {age}'),
        ('user', 'How are you?')
    ]
)

prompt_five_year = prompt_tempelate.invoke(
    {
        'age':'5 years',
    }
)

prompt_eighty_year = prompt_tempelate.invoke(
    {
        'age':'80 years',
    }
)

llm.invoke(prompt_five_year).content
llm.invoke(prompt_eighty_year).content


```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI(model = 'gpt-5', temperature = 0)

prompt_tempelate = ChatPromptTemplate.from_messages(
    [
        ('system', 'You are a cool assistant, whose age is {age}'),
        ('user', 'How are you?')
    ]
)

chain = prompt_tempelate | llm

chain.invoke(
    {
        'age' : '500 years'
    }
) 
```
